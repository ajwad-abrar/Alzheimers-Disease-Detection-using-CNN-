```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from zipfile import ZipFile
file_name = "Alzheimers-ADNI.zip"

with ZipFile(file_name,'r') as zip:
  zip.extractall()
  print('done')
```

```python
images = []
lables = []

## AD Patient Brain Scan Image ##
for i in range(1, 172):
    img = cv2.imread(f"Alzheimers-ADNI/train/Final AD JPEG/AD ({i}).jpg")
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    gray = gray/255
    gray = cv2.resize(gray,(200, 200))
    images.append(gray)
    lables.append(0)
```

```python
plt.figure(figsize = (15,15))
for i in range(20):
    plt.subplot(4, 5, i + 1)
    plt.imshow(images[10 + i*3])
plt.show()
```

```python
## CN Patient Brain Scan with PreProcessing of the Image ##
for i in range(1, 581):
    img = cv2.imread(f"Alzheimers-ADNI/train/Final CN JPEG/CN ({i}).jpg")
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    gray = gray/255
    gray = cv2.resize(gray,(200, 200))
    images.append(gray)
    lables.append(1)  # CN
```

```python
# Sample Images
plt.figure(figsize = (15,15))
for i in range(20):
    plt.subplot(4, 5, i + 1)
    plt.imshow(images[200 + i*3])
plt.show()
```
